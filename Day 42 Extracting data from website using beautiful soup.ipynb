```python
# collecting first page of artisits' list

page = requests.get('https://web.archive.org/web/20121007172938/http://www.nga.gov/collection/anK1.htm')

# creating a BeautifulSoup object

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
# Pulling all text from the BodyText div

artist_name_list = soup.find(class_='BodyText')

# Pulling text from all instances of <k> tag within BodyText div
artist_name_list_items = artist_name_list.find_all('a')
```

```python
# creating for loop to print out all artists' name using prettify() method

for artist_name in artist_name_list_items:
    print(artist_name.prettify())
```

#### Using BeautifulSoup to select particular content

Since our interest is to find names and URLs of the various member of 43rd Congress, we can see that the the content of our interest in closed in '< a >', '< /a >'. Hence, we will use beautifulsoup to extract name of congress and their URLs as follows:

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(open('43-congress-beautifulsoup'), features = 'lxml')

links = soup.find_all('a')

for link in links:
    print(link)
```
