### An algorithm for discovering Lagrangians automatically from data

(Paper link: https://peerj.com/articles/cs-31.pdf)'

#### Abstract:
This paper presents an algorithm that automates the model building process in a scientifically principled way.

The algorithm can take observed trajectories from a wide variety of mechanical systems and, without any other prior knowledge or tuning of parameters, predict the future evolution of the system. It does so by applying the principle of least action and searching for the simplest Lagrangian that describes the system's behaviour. By generating this Lagrangian in a human interpretable form, it can also provide insight into the workings of the system.

### The Principle of Least Action:

This is one of the most fundamental and most celebrated principle in physics.

First proposed by Maupertuis (1744)and Euler (1744) it states that the problem of predicting the behaviour of many physical systems can be cast as finding the behaviour that minimises the expenditure given by some cost function. The total expenditure of thesystem is known as the action. It is a remarkable fact that the behaviour of a very wide rangeof physical systems—including those studied in classical mechanics, special and generalrelativity, quantum field theory, and optics—can all have their behaviour explained interms of minimising a cost function.

Each physical system has its own cost function, and once this function is known it is possible to predict exactly what the system will do in the future. The exercise of determining the cost function-often known as Lagrange's function. 

#### Advantages of Lagrangian for automated modelling algorithm:

1. It is a single, scalar expression that contains everything necessary to predict the system's future evolution. Meantime, in contrast to it, in Hamiltonian, it is necessary to find the corresponding conjugate momenta. 

2. It is coordinate system independent. However, it should be noted that not all cases can be described by the Lagrangian like dissipative systems.


### The Algorithm

The algorithm fundamentally consists of three elements:

1. Objective to guide the search; Score Function
2. Representation
3. Search of possible Lagrangian

1. Score Function:

The objective of this algorithm is to find a Lagrangian that, when integrated along any neighbouring trajectory, yields a smaller total (action) than when integrated along any neighbouring trajectory.



(Image source: Paper itself!)

A Normalisation score for each candidate Lagrangian is introduced to make sure that the minimized Euler-Lagrange (as shown in above figure) is the desired Lagrangian. Further, there is another case where minimised Euler-Lagrange (EL) is not the desired Lagrangian. This case is more interesting because this Lagrangian satisfies the Euler-Lagrange equations trivially, in a way that does not depend on the trajectory, C. Hence, author reasoned that the desired Lagrangian should perform better in EL_D but should score poorly on EL_C (in controlled trajectory). 

2. Representation and search

Two representations of candidate Lagrangians: 

* Restricted polynomial representaion (allows a fast search algorithm to be implemented.It is limited to Lagrangians it can represent exactly, although through Taylor's theorem it can find approximations to any Lagrangian.

* Generalisation

The restricted polynomial representation assumes that the Lagrangian can be represented by a polynomial in the coordinates and velocities. The model is a sum of monomial terms, parameterised by coefficients multiplying every term. The polynomial was restricted in two ways: limit the maximum power of any coordinate or velocity to be m; and limit the maximum degree of any combination of coordinates and velocities to p. Further, any terms from the model that can have no physical significance were removed.
