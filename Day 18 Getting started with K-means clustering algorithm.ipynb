```python
#importing the necessary libraries

from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
Data = {'x': [25,34,22,27,33,33,31,22,35,34,67,54,57,43,50,57,59,52,65,47,49,48,35,33,44,45,38,43,51,46],
        'y': [79,51,53,78,59,74,73,57,69,75,51,32,40,47,53,36,35,58,59,50,25,20,14,12,20,5,29,27,8,7]
       }
```

```python
df = DataFrame(Data, columns = ['x', 'y'])
```

```python
plt.scatter(df['x'], df['y'])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3).fit(df) # three clusters
centroids = kmeans.cluster_centers_
print ('centroids: \n \n', centroids)
```

```python
#visualizing the clusters
plt.scatter(df['x'], df['y'], c = kmeans.labels_.astype(float), s = 50, alpha = 0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], marker = 'o', c = 'g', s = 50)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# into 4 clusters 

kmeans = KMeans(n_clusters = 4).fit(df) # three clusters
centroids = kmeans.cluster_centers_
print ('centroids: \n \n', centroids)

#visualizing the clusters
plt.scatter(df['x'], df['y'], c = kmeans.labels_.astype(float), s = 50, alpha = 0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], marker = 'o', c = 'g', s = 50)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#visualization using tkinter
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
```

```python
root = tk.Tk()

canvas1 = tk.Canvas(root, width = 100, height = 100)
canvas1.pack()

label1 = tk.Label(root, text = centroids, justify = 'center')
canvas1.create_window(70, 50, window=label1)

figure1 = plt.Figure(figsize = (5, 4), dpi = 100)
ax1 = figure1.add_subplot(111)
ax1.scatter(df['x'], df['y'], c = kmeans.labels_.astype(float), s = 50, alpha = 0.5)
ax1.scatter(centroids[:, 0], centroids[:, 1], c = 'g', s = 50)
scatter1 = FigureCanvasTkAgg(figure1, root)
scatter1.get_tk_widget().pack(side=tk.LEFT, fill = tk.BOTH)

root.mainloop()
```

```python
# trying k-means for multiple features/variables

import pandas as pd
df = pd.read_csv('Heart Fail Prediction.csv')
print(df.head())
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2).fit(df)
centriods = kmeans.cluster_centers_
print('centriods: \n \n', centriods)
```

```python
#visualizing k means
import matplotlib.pyplot as plt

predict = kmeans.predict(df)
df['cluster'] = predict

pd.plotting.parallel_coordinates(df, 'cluster')
plt.rcParams['figure.figsize'] = (30, 30)
```
