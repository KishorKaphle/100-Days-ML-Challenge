# Fast unfolding of communities in large networks

Source: https://arxiv.org/pdf/0803.0476.pdf

## Abstract:

1. Heuristic method that works based on modularity optimization
2. Better in terms of computational time

## Performance of the mode is measured in terms of modularity of the partition:
It is a scalar value which measures the density of links inside communities as compared to links between communities and lies in between -1 and 1. (Like Silhouette analysis?)




## How it works?

This method basically has twp phases which is continuously repeated during the operation. 





Louvian Community Detection Algorithm is a greedy optimization method that apparently execute in time O(nlog2^n) in the number of nodes in the network.

This method has been used with success for network of many different type and for sizes up to 100 million nodes and billions of links. Exact modularity optimization is known to be NP-hard.

(Source :https://perso.uclouvain.be/vincent.blondel/research/louvain.html)

```python
# importing the required libraries

#!pip install python-louvain

import networkx as nx
from community import community_louvain
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
# defining the graph

edge = [(1, 2), (1, 3), (1, 5), (1, 6), (2, 7), (2, 8), (2, 9)]
G = nx.Graph()
G.add_edges_from(edge)
```

```python
def draw_graph(g):
        pos = nx.spring_layout(g)
        nx.draw_networkx_nodes(g, pos)
        nx.draw_networkx_labels(g, pos)
        nx.draw_networkx_edges(g, pos, width = 1.0, alpha = 0.5)

        graph = nx.Graph()

        graph.add_edges_from(edge)
                      

draw_graph(G)
```

```python
# returning partition as a dict
partition = community_louvain.best_partition(G)
```

```python
# draw the graph
draw_graph(G) # initial graph

plt.title('Original graph')

plt.show()

pos = nx.spring_layout(G)

# coloring the node according to their partition
cmap = cm.get_cmap('viridis', max(partition.values()) +1)

nx.draw_networkx_nodes(G, pos, partition.keys(), node_size = 100, cmap = cmap, node_color = list(partition.values()))

nx.draw_networkx_labels(G, pos)

nx.draw_networkx_edges(G, pos, alpha = 0.5, width = 1.0)

plt.title('Partitioned graph')

plt.show()
```
