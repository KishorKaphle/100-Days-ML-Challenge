## Final model for getting Adjacency matrix

```python
#Creating Adjacency Matrix
import numpy as np

graph ={"1" : ["2", "3"],
       "2" : ["1", "4"],
       "3" : ["1", "4"],
       "4" : ["5", "2"],
       "5" : ["4", "3"]}

def AdjMat(graph):
    
   
    l = list()
    
    count = 0
    
    for key in graph:
        count+=1
        count1 =1
        for item in graph.values():
            if count == count1:
                    k = str (key)
                    i0 = str(item[0])
                    i1 = str(item[1])
            
                    a = k +i0
                    b = k + i1
                    l.append(a)
                    l.append(b)
            count1+=1
                    
        #print(l)
    
  
    size = []
    
    for key in graph:
        size.append(key)
        n = len (size)
    
    #print ('dimension of A matrix', n, n)
    
    # creating quare zero matrix of size n x n
    A = numpy.zeros((n,n))
    
    #print (A)
    
    for j in l:
        a=0
        a = [int(i) for i in j]
        #print((a[0]-1),(a[1]-1))
        A[a[0]-1][a[1]-1]=1

    #print('Adjacency Matrix:\n \n', A)
    
    
    return A
        
```

```python
#calling function for adjacency matrix
print (graph)

AdjMat(graph)
```

## Laplacian Matrix, L = D-A

```python
graph ={"1" : ["2", "3"],
       "2" : ["1", "4"],
       "3" : ["1", "4"],
       "4" : ["5", "2"],
       "5" : ["4", "3"]}

def lapMat(graph):
    L = degreeMat (graph)-AdjMat(graph)
    
    return L

print ("Laplacian Matrix: \n \n", lapMat(graph))
    
```

```python
graph ={"1" : ["2", "3"],
       "2" : ["1", "4"],
       "3" : ["1", "4"],
       "4" : ["5", "2"],
       "5" : ["4", "3"]}

def eign(L):
    from numpy.linalg import eig

    values, vectors = eig(L) # selected only V not E because our main goal is to work on vertices, after all E can be easily retained

    #print ("eigen values: ", values, "\n \n", "eigen vectors: ", vectors)

    return (values, vectors)

print ('eigen values:\n \n', eign(lapMat(graph))[0])

print ('\n \n eigen vectors:\n \n', eign(lapMat(graph))[1])
```

```python
import pandas as pd

#noOfClusters = 2, say
#k_neighbors  = 3, say 



def spectral(noOfClusters, k_neighbors):
    
    from sklearn.cluster import KMeans
    import matplotlib.pyplot as plt

    eigenValues, eigenVectors = eign(lapMat(graph))
    sortedEigenValueIndex =  np.argsort(eigenValues.real)
    sortedEigenVectors = eigenVectors[:,sortedEigenValueIndex].real
    sortedEigenVectors = sortedEigenVectors[:,:noOfClusters]
    eigenDataFrame = pd.DataFrame(sortedEigenVectors)
    columnNames = eigenDataFrame.columns.values
    columnNames = [str(x) for x in columnNames]
    eigenDataFrame.columns = columnNames
    
    print (eigenDataFrame.iloc[:, 0])
    
    print (eigenDataFrame.iloc[:, 1])
    
    

    kmeans = KMeans(n_clusters = noOfClusters).fit(eigenDataFrame) # three clusters
    centroids = kmeans.cluster_centers_
    print ('centroids: \n \n', centroids)
    
    
    
    plt.scatter(eigenDataFrame.iloc[:, 0],eigenDataFrame.iloc[:, 1], c = kmeans.labels_.astype(float), s = 50, alpha = 0.5)
    plt.scatter(centroids[0], centroids[1],c = 'g', marker = '*', s = 50)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    
    
        
    return (eigenDataFrame)
```

```python


# defining function to visualize the graphs
import networkx as nx
graph ={"1" : ["2", "3"],
       "2" : ["1", "4"],
       "3" : ["1", "4"],
       "4" : ["5", "2"],
       "5" : ["4", "3"]}

def getNodeArray(graph):

    #seperating edges from the graph dict
    g = []
    count =0
    for key in graph:
            count+=1
            count1 =1
            for item in graph.values():
                if count == count1:
                    k = int(key)
                    i0 = int(item[0])
                    i1 = int(item[1])
                    g.append((k, i0))
                    g.append((k, i1))
               
                count1+=1
    return g


g = getNodeArray(graph)


def graphShow(g):


    def draw_graph(g):
        pos = nx.spring_layout(g)
        nx.draw_networkx_nodes(g, pos)
        nx.draw_networkx_labels(g, pos)
        nx.draw_networkx_edges(g, pos, width = 1.0, alpha = 0.5)

    graph = nx.Graph()

    graph.add_edges_from(g)
                      
    draw_graph(graph)
    
    return
    
graphShow(g)
```
