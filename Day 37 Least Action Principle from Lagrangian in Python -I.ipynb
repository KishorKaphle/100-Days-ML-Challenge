```python
x = df['Interest_Rate']
y = df['Unemployment_Rate']
z = df['Stock_Index_Price']
print(x[0])
```

```python
def adders(x, y, z):
    add =[]
    for i in range(len(z)):
        a = z[i]
        b = (x[i]+y[i])
        c = a-b
        
        #print (a, b)
        #print(c)
        add.append(c)
    
    print (add)
        
    
    
    return add
```

```python
d = adders(x, y, z)
```

```python
import matplotlib.pyplot as plt


plt.plot(z, c = 'b', label = 'output', marker = '*')

plt.plot(d, c = 'r', label = 'adders', marker ='o')

plt.legend()
plt.show()
```

```python

def dynamics(d):
    base_adder = d[0] # let's say the initial condition to be intial stage of the system dynamics and every offsets comming in series to be the changes in the dynammics
    mechanics = []
    for i in range (len(d)-1):
        e = d[i+1]-d[i];
        mechanics.append(e)
    
    #print (mechanics)
    
    return mechanics
```

```python
dynamics(d)
```

```python
import matplotlib.pyplot as plt

base_adder = d[0]

plt.plot(z, c = 'b', label = 'output', marker = '*')

plt.plot(base_adder, c ='r', label = 'base adder', marker = '+')

plt.plot(dynamics(d), c = 'g', label = 'dynamics', marker ='*')

plt.legend()
plt.show()
```
