For colored picture, array is of 3 dimensional (2D image and 1D color grade of major color R, B and G. On the other hand, for black & white image, picture is interpreted as an array of 2D only.

Let's import image from url

```python
import numpy as np
import urllib.request
import cv2


def image_from_url(url):
    gett = urllib.request.urlopen(url)  # it opens the url to read the content

    img = np.asarray(bytearray(gett.read()), dtype='uint8')
    img = cv2.imdecode(img, cv2.IMREAD_COLOR)

    return img


image = image_from_url('https://external-content.duckduckgo.com/iu/?u=http%3A%2F%2Fsasscer.files.wordpress.com%2F2014%2F03%2Funique-characteristics-giant-panda.jpg&f=1&nofb=1')


cv2.imshow('Image', image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
